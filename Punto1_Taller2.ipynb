## 1. Raices de polinomios

### 1.1. ¿De qué tipo es el error asociado a la estimación de raíces usando el método de Newton-Raphson?

El error asociado al método de Newton-Raphson es un error de truncamiento. Un error de truncamiento ocurre cuando una iteración que requiere un número infinito de pasos para alcanzar el valor real, se detiene en un número finito de pasos; lo que sucede con el método referido.

### 1.2. ¿Cómo ajustar la precisión para estimar raíces con el método de Newton-Raphson?

La precisión se ajusta mediante el valor establecido para el error. Una mayor precisión significa un error más pequeño. Una vez que se determina el error, debe iterarse hasta que se alcance el valor deseado para el mismo. Un error más pequeño significa una distancia más pequeña entre Xk+1 y Xk, y se sabe que esa distancia se reduce con el número de iteraciones. Entonces, si se quiere una alta precisión el número de iteraciones debe ser grande.
